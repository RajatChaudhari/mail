```python
mail=pd.read_csv('Rajat_Chaudhari.csv')
```

```python
mail=mail.rename(columns = {'Unnamed: 0':'SrNo'})
```

```python
sep="regards"
mail["Message"]=mail["Message"].apply(lambda x: (str(x).lower()).split(sep, 1)[0])
```

```python
mail["SubLen"]=mail["Subject"].apply(lambda x: len(str(x)))
mail["MsgLen"]=mail["Message"].apply(lambda x: len(str(x)))
```

```python
nltk.download('punkt')
```

```python
test=nltk.word_tokenize(mail["Message"][0])
```

```python
mail["MsgTokens"]=mail["Message"].apply(lambda x:nltk.word_tokenize(str(x)))
```

```python
mail.head(2)
```

```python
mail["MsgSentenceTokens"]=mail["Message"].apply(lambda x:nltk.sent_tokenize(str(x)))
```

```python
mail["MsgSentenceTokens"][0]
```

```python
def normalize(x):
    norm=[word.lower() for word in x if word.isalpha()]
    return norm
```

```python
mail["MsgTokens"]=mail["MsgTokens"].apply(lambda x: normalize(x))
#removed the digit 2 as well
```

```python
mail["MsgTokens"][0]
```

```python
nltk.download('averaged_perceptron_tagger')
#nltk.help.upenn_tagset()
```

```python
mail["POS_Msg"]=mail["MsgTokens"].apply(lambda x:nltk.pos_tag(x))
```

```python
nltk.download('universal_tagset')
```

```python
mail["POS_Msg_Uni"]=mail["MsgTokens"].apply(lambda x:nltk.pos_tag(x,tagset='universal'))
```

```python
def noun_POS(x):
    words=nltk.pos_tag(x,tagset='universal')
    mw= [word[0] for word in words if word[1]=='NOUN']
    return mw
    #also check what is pos tag sentences
```

```python
mail["POS_Msg_Noun"]=mail["MsgTokens"].apply(lambda x:noun_POS(x))
```

```python
mail["POS_Msg"][0]
```

```python
mail["POS_Msg_Noun"][0]
```

```python
nltk.download('maxent_ne_chunker')
```

```python
nltk.download('words')
```

```python
def chunkMsg(x):
    entities=nltk.chunk.ne_chunk(x)
    return x
```

```python
mail["Chunked"] = mail["POS_Msg"].apply(lambda x: chunkMsg(x))
```

#  Extracting text from webpages

```python
import nltk
from urllib.request import urlopen
```

```python
url="http://scikit-learn.org/stable/tutorial/basic/tutorial.html#learning-and-predicting"
```

```python
html=urlopen(url).read()
```

```python
html
```

```python
from bs4 import BeautifulSoup
```

```python
web_str=BeautifulSoup(html).get_text()
```

```python
web_str
```

```python
start=web_str.find("In general, a learning problem")
```

```python
end=web_str.find(" labels for each instance.")
```

```python
last_sent=len("As a result, predict() returns a 2d array with multiple predicted labels for each instance.")
```

```python
intro=web_str[start:end+last_sent]
```

```python
intro_token=nltk.word_tokenize(intro)
```
